```python
from google_play_scraper import app, reviews, Sort
import pandas as pd
import csv
```

# ***TAHAP 2 ‚Äî Tentukan ID Aplikasi***

```python
# ID aplikasi (contoh: https://play.google.com/store/apps/details?id=com.larus.wolf&hl=id)
app_id = 'com.larus.wolf'
```

# ***TAHAP 3 ‚Äî Fungsi Mengambil Ulasan***

```python
def get_review(app_id, lang='id', count=80000, sort=Sort.NEWEST):
    try:
        result, continuation_token = reviews(
            app_id,
            lang=lang,
            country='id',
            sort=sort,
            count=count
        )
        return result
    except Exception as e:
        print("‚ùå Error:", e)
        return None
```

# ***TAHAP 4 ‚Äî Fungsi Simpan ke CSV & Excel***

```python
def export_reviews(reviews_data, base_filename='hasil_scraper_ulasan_APP'):
    if reviews_data:
        # Buat DataFrame
        df = pd.DataFrame(reviews_data)[['reviewId', 'userName', 'score', 'content', 'at']]
        df.columns = ['reviewId', 'userName', 'rating', 'review_txt', 'date']

        # Simpan ke CSV
        csv_filename = f'{base_filename}.csv'
        df.to_csv(csv_filename, index=False, encoding='utf-8')
        print(f"‚úÖ Data berhasil disimpan ke file CSV: {csv_filename}")

        # Simpan ke Excel
        excel_filename = f'{base_filename}.xlsx'
        df.to_excel(excel_filename, index=False, engine='openpyxl')
        print(f"‚úÖ Data berhasil disimpan ke file Excel: {excel_filename}")

        return df
    else:
        print("‚ö†Ô∏è Tidak ada data untuk disimpan.")
        return None
```

# ***TAHAP 5 ‚Äî Jalankan Scraper***

```python
reviews_data = get_review(app_id)

if reviews_data:
    df = export_reviews(reviews_data, 'hasil_scraper_ulasan_APP')
    print(f"\nJumlah total ulasan: {len(df)}\n")
else:
    print("Tidak ada ulasan ditemukan.")
```
